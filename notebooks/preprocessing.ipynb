```python
import os
import pillow_avif
from PIL import Image
```

```python
# 1. 경로 설정 
# 예: dataset/images/day, dataset/images/night
base_dir = "../dataset/images" 
categories = ["day", "night"]

# 2. 전처리 함수 정의
def convert_images_to_jpg(directory):
    print(f"Processing directory: {directory}")
    for filename in os.listdir(directory):
        if filename.lower().endswith(('.png', '.webp', '.avif', '.jpeg')):
            filepath = os.path.join(directory, filename)
            
            try:
                # 이미지 열기
                img = Image.open(filepath).convert("RGB") # 투명도(RGBA) 등을 RGB로 통일
                
                # jpg 이름 만들기 (확장자만 변경)
                new_filename = os.path.splitext(filename)[0] + ".jpg"
                new_filepath = os.path.join(directory, new_filename)
                
                # 저장 및 원본 삭제 (선택사항)
                img.save(new_filepath, "JPEG", quality=90)
                if filename != new_filename: # 원본이 jpg가 아니면 삭제
                    os.remove(filepath)
                    print(f"Converted & Removed: {filename} -> {new_filename}")
                    
            except Exception as e:
                print(f"Error converting {filename}: {e}")

# 3. 실행
for category in categories:
    path = os.path.join(base_dir, category)
    convert_images_to_jpg(path)

print("All images converted to JPG!")
```
